```julia
using PyPlot
```

```julia
circles_top(x, r, n, m) = sqrt(r^2 - (x - n)^2) + m
circles_bottom(x, r, n, m) = - sqrt(r^2 - (x - n)^2) + m
```

```julia
N = 100
r = 0.1
for n = -30:5
    for m = -10:30
        for i = 1:N-1
            x = 2r*i/N + n - r
            points_top = circles_top(x, r, n, m)
            points_bottom = circles_bottom(x, r, n, m)
            plt.plot(x, points_top, "k.", markersize=1, alpha=0.1)
            plt.plot(x, points_bottom, "k.", markersize=1, alpha=0.1)
        end
        plt.plot(n - r, m, "k.", markersize = 1, alpha = 0.1)
        plt.plot(n + r, m, "k.", markersize = 1, alpha = 0.1)
    end
end
plt.axis("equal")

coordsx = Real[]
coordsy = Real[]
for i = 1:length(coords)
    push!(coordsx, coords[i][1])
    push!(coordsy, coords[i][2])
end
plt.plot(coordsx, coordsy, "b", markersize=1, alpha=1)
```
